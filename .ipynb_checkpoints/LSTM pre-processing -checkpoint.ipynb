```python
import pandas as pd
import numpy as np
import tensorflow as tf 
import pickle
from sklearn.metrics import mean_squared_error
import seaborn as sns 
import matplotlib.pyplot as plt
from numpy.lib.stride_tricks import sliding_window_view
import datetime as dt
import random
import os 
curr_dir = 'C:\\Users\\Zain Ahmed\\Desktop\\Literature review'

os.chdir(curr_dir)

data = pd.read_table('SCiBER.txt',delimiter='\t')

# Dropping duplicates

data.drop_duplicates(subset=['Date'],keep='last',inplace=True)
data.groupby(data['Date']).size().sort_values(ascending=False)

# Converting the date column to datetimeindex and converting data to float32

data.set_index(pd.to_datetime(data['Date'],format='%d.%m.%Y %H:%M'),inplace=True)
data

# Ensuring Data is continous 

data = data.asfreq(freq='15min')

# Filtering only office buildings 

office_data=data.filter(regex='Office')

# Dividing based on missing data (data availability more than 99.9%)

office_filter = (office_data.isnull().sum()*100/len(office_data)).sort_values(ascending=False)

office_filter = office_filter[office_filter<0.1]

office_filter

office_data_filtered = office_data.filter(items=office_filter.index,axis=1)
```

# Dividing data in train, validation and test set 

```python
train_rows = int((office_data_filtered.shape[0]*0.70)/96)*96
val_rows = int((office_data_filtered.shape[0]*0.90)/96)*96
train = office_data_filtered[:train_rows]
vali = office_data_filtered[train_rows:val_rows]
test = office_data_filtered[val_rows:]
```

# We will remove the weekends and keep the time betwwen 5 am and 7 pm 
# One day consists of 56 timesteps as 
# (4 timesteps/hour x 14 hours a day)



```python
def pre_process(of_da_fi):
    '''
    Takes a dataframe and 
    
    - keeps only data between 5am and 7pm
        
    - replaces NA values by forward filling
    
    - performes normalization (value-mean)/std

    '''
    of_da_fi = of_da_fi[(of_da_fi.index.dayofweek<5) & ((of_da_fi.index.hour>4) & (of_da_fi.index.hour<19))]
    
    of_da_fi = of_da_fi.fillna(method='ffill')
    
    check_na = (of_da_fi.isnull().sum()*100/len(of_da_fi)).sort_values(ascending=False)
    
    plt.figure(figsize=(15,15))
    
    sns.violinplot(data=of_da_fi)
    
    plt.xticks(rotation=90)
    
    of_da_fi = (of_da_fi-of_da_fi.mean())/of_da_fi.std()
    
    return of_da_fi

train_data = pre_process(train)
vali_data = pre_process(vali)
test_data = pre_process(test)
```

# Rolling window function 

```python
def rolling_window(array_data,input_size,output_size,offset):
    '''
    The function takes a series of size (array_data,1) and generates 
    
    a array of size (input_size,((array_size-(input_size+output_size)/stride)+1)) as inputs 
    
    and array of size (array_size-(input_size+output_size)/stride)+1)as output 
    
    '''
    
    data = sliding_window_view(array_data.to_numpy(),window_shape=input_size)[::offset][:-1]
        
    labels = sliding_window_view(array_data.to_numpy()[input_size:],window_shape=output_size)[::offset]
    
    assert data.shape[0] == labels.shape[0],"Training rows and labels are unequal"    
    
    if output_size == offset:
        
        ra_in = random.randint(0,data.shape[0]-2) 
        
        a = data[ra_in+1,-offset:]
        
        b = labels[ra_in]
        
        np.alltrue(a == b)
    
    data = np.expand_dims(data,axis=2)
    
    labels = np.expand_dims(labels,axis=2)
        
    return data,labels
```

# Office on column 10 is used here to generate rolling window and corresponding labels 

```python
tra_data_sli,tra_lab = rolling_window(train_data.iloc[:,10],input_size=280,output_size=56,offset=56)
val_data_sli,val_lab = rolling_window(vali_data.iloc[:,10],input_size=280,output_size=56,offset=56)
tes_data_sli,tes_lab = rolling_window(test_data.iloc[:,10],input_size=280,output_size=56,offset=56)
```

```python
np.save('tra_data_sli',tra_data_sli)
np.save('val_data_sli',val_data_sli)
np.save('tes_data_sli',tes_data_sli)
np.save('tra_lab',tra_lab)
np.save('val_lab',val_lab)
np.save('tes_lab',tes_lab)
```
